# Exploratory analysis (example)
This notebook shows how to load the synthetic data and reproduce the PCA plot.


```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
df = pd.read_csv('data/processed/expression_synthetic.csv', index_col=0)
X = df.T
pca = PCA(n_components=2)
pcs = pca.fit_transform(X - X.mean(axis=0))
plt.scatter(pcs[:,0], pcs[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of samples (synthetic)')
plt.show()
```
